# Matplotlib

### Linecharts

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
##linspace is the method to generate numpy array, 
##it requires 3 arguments, startpoint, endpoint and number of points in between
x = np.linspace(0, 10,10)
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
##adding extra features to the plot
plt.plot(x,y)
plt.xlabel("Time")
plt.ylabel("Sin Function")
plt.title("My Cool Chart")
plt.show()
```
